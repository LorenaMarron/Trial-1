```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,2,0.01)

#print(t)

f = 1

y = np.cos(2*np.pi*f*t)
plt.plot(t,y)

print(t.shape)

z = np.random.normal(0,0.1,t.shape[0])
#plt.plot(t,z)

y_noisy = y + z
plt.plot(t,y_noisy)

from scipy import signal

b, a = signal.butter(5,0.1,'low')
y_filtered = signal.filtfilt(b,a,y_noisy)
plt.plot(t,y_filtered)
```
